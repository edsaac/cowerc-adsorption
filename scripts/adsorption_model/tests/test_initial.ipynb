```python
parameters = ThomasModelParameters(k_T=0.1, q_m=0.2, b=0.3)
setup = ThomasExperimentalSetup(
    C_0=0.4, length=1, pore_velocity=0.5, rho_p=0.6, epsilon=0.7
)
model = ThomasModel("test")
```

```python
exp = Experiment(
    "my-experiment", "my-contaminant", model=model, setup=setup, parameters=parameters
)
```

```python
import numpy as np

t = np.arange(0, 101, 10)
c = 0.5 * t
exp.btc = BreaktroughData(time=t, conc=c)
exp.setup.C_0 = 100
```
