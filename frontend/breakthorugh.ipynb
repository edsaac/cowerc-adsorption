```python
from organize_foams.boundary_probe import boundaryProbe
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
organized_data_folder = Path("../tests/first_order_adsorption/template/organizedData/")
```

```python
probes = boundaryProbe(
    organized_data_folder/"points_Baq_Caq.xy",
    organized_data_folder/"time.txt",
    organized_data_folder/"xyz.txt"
)
```

```python
data = probes.array_data
```

```python
data.time
```

```python
fig, ax = plt.subplots()

for k, v in data.items():
    if k == "Baq":
        ax.plot(v.time, v[0],
                label=f"Outflux {k}")
        ax.plot(v.time, v[1],
                label=f"Influx {k}")

ax.set_xlabel("Time [s]")
ax.set_ylabel("Concentration [kg/m³]")
ax.legend()
plt.show()
```
