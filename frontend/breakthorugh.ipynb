```python
from organize_foams.boundary_probe import boundaryProbe
from pathlib import Path
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
```

```python
cmap = mpl.cm.Dark2
colors = cmap(np.linspace(0, 1, 3))

def plot_data(data):
    fig, ax = plt.subplots()

    for k, v in data.items():
        
        match k:
            case "Caq"|"Cads":
                color = colors[0]
            case "Baq"|"Bads":
                color = colors[1]

        ax.plot(v.time, v[0],
                label=f"Outflux {k}", c=color, lw=3)
        ax.plot(v.time, v[1],
                label=f"Influx {k}", c=color, ls="dashed")

    ax.set_xlabel("Time [s]")
    ax.set_ylabel("Concentration [kg/m³]")
    ax.legend()
    plt.show()
    
```

```python
organized_data_folder = Path("../tests/first_order_adsorption/case/organizedData/")
```

```python
probes = boundaryProbe(
    organized_data_folder/"points_Baq_Caq.xy",
    organized_data_folder/"time.txt",
    organized_data_folder/"xyz.txt"
)

data = probes.array_data
```

```python
data.time
```

```python
plot_data(data)
```
